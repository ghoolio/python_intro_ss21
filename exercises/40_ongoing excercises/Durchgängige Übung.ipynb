## Funktionen
Dieser Teil befasst sich nicht mit der Implementierung von weiteren Funktionalitäten, sondern damit Ihr Programm übersichtlicher zu gestalten. Hierzu sollen Sie verschiedene Funktionen erstellen, die von einer Hauptfunktion aufgerufen werden. Definieren Sie dazu folgende Funktionen:

1.	Main / Diese Funktion rufen Sie zu Beginn auf und nutzen diese als „Drehpunkt“ Ihres Programmes
2.	Game / Diese Funktion soll ein neues Spiel starten, wie Sie es zuvor entworfen haben
3.	Calculate/ Diese Funktion nimmt den Eingabe Wert des Benutzers und die Höhe entgegen und berechnet daraus die neue Höhe
4.	CheckHeight / Diese Funktion nimmt den Parameter Höhe nach einer Berechnung entgegen und überprüft, ob ein Minimum oder Maximum überschritten bzw. unterschritten wurde. Diese Funktion soll wahr zurückgeben, falls alles ok ist und Falsch, wenn die Grenzen überschritten wurden

Sollte der Benutzer verlieren, geben Sie dem Benutzer über die Main Funktion die Möglichkeit ein neues Spiel zu starten, oder das Programm zu beenden. Hierbei sollen nur Funktionsaufrufe und keine Schleifen verwendet werden.

Um diese Übung abzuschließen, sollte Ihr Programm nun folgendes können:

1.	Über die Main Funktion das Spiel steuern
2.	Über die Game Funktion eine einzelne „Session“ steuern
3.	Die Calculate, CheckHeight und Write Funktion benutzen, um das Spiel zu modularisieren

5.	Alle Anforderungen aus dem Teil Variablen und Datentypen

7.	Alle Anforderungen aus dem Teil Komplexe Datentypen
8.	Alle Anforderungen aus dem Teil Schleifen



```python
def CheckHeight(Height):
    MaxHeight = 5
    MinHeight = 0
    if(Height < MaxHeight and Height > MinHeight):
        return True
    else:
        return False

def Calculate(userinput, Height):
    return Height + int(userinput) * 0.5 - 2

def Game():
    Height = 2
    Output = []
    while True:
        Clicks = input("Anzahl Klicks?")
        Height = Calculate(Clicks,Height)
        if(CheckHeight(Height) == True):
            Temp = ["-","-","-","-","-","-","-","-","-","-","-"]
            Position = int(Height * 2)
            Temp[Position] = "|"
            Output.append(Temp)
            Write(Output)
        else:
            print("Verloren! Die aktualisierte Höhe ist außerhalb des Spielfeldes.")
            return

def Write(List):
    Length = len(List)
    for x in range(0,11):
        Ergebnis = []
        for element in range(0,Length):
            Ergebnis.append(List[element][10 - x])
        print(Ergebnis)

def Main():
    Game()
    Restart = input("Für eine neue runde drücken Sie 1, um das Programm zu beenden geben Sie 0 ein")
    if(int(Restart) == 1):
        Main()

Main()
```

```python
import random

def CheckHeight(Height):
    with open("Landschaft.csv","r") as File:
        lines = File.readlines()
        if lines:
            last_line = lines[-1]
            last_line = last_line.replace(";","").lstrip().split()
    MaxHeight = float(last_line[0])
    MinHeight = float(last_line[1])
    if(Height < MaxHeight and Height > MinHeight):
        return True
    else:
        return False

def Calculate(userinput, Height):
    return Height + int(userinput) * 0.5 - 2 + random.randrange(-1,2)* 0.5

def CreateField():
    with open("Landschaft.csv", "a") as Field:
        HNO = 5 - (random.randrange(1,4) * 0.5)
        HNU = 0 + (random.randrange(1,4) * 0.5)
        Field.write("\n " + str(HNO) + ";" + " "+ str(HNU) + ";")
        Field.close()
        return HNO,HNU
def Game():
    Height = 2
    Output = []
    while True:
        Clicks = input("Anzahl Klicks?")
        ERG = CreateField()
        Height = Calculate(Clicks,Height)
        if(CheckHeight(Height) == True):
            Temp = ["-","-","-","-","-","-","-","-","-","-","-"]
            Position = int(Height * 2)
            Temp[Position] = "|"
            for value in range(int(ERG[0] * 2),11):
                Temp[value] = "/"
            for value in range(0,int(ERG[1] * 2)):
                Temp[value] = "/"
            Output.append(Temp)
            Write(Output)
        else:
            print("Verloren! Die aktualisierte Höhe ist außerhalb des Spielfeldes.")
            return
        
def Write(List):
    Length = len(List)
    with open('Output.txt', 'w') as o:
        for x in range(0,11):
            Ergebnis = []
            o.write(str((10 - x)/2) + "|")
            for element in range(0,Length):
                o.write(List[element][10 - x])
            o.write("\n")
        o.write(str(Length) + " Ist der Score")
    o.close()
    for x in range(0,11):
        Ergebnis = []
        for element in range(0,Length):
            Ergebnis.append(List[element][10 - x])
        print(Ergebnis)

def Main():
    Game()
    Restart = input("Für eine neue runde drücken Sie 1, um das Programm zu beenden geben Sie 0 ein")
    if(int(Restart) == 1):
        Main()

Main()
```
